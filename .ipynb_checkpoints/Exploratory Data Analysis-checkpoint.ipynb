```python
import pandas as pd
```

```python
groceries = pd.read_pickle("./Data/groceries_with_new_columns.pkl")
groceries.head()
```

```python
groceries[groceries['Low Inventory'] == 'Low Inventory']
```

```python
groceries['Subcategory'] = groceries['Subcategory'].str[1:]
## also do (.str.strip())
```

```python
groceries[groceries['Subcategory'] == "Snacks"]
```

```python
mask = (groceries['Low Inventory'] == 'Low Inventory') & (groceries['Subcategory']=='Snacks')
groceries.loc[mask,['Subcategory','Item','Inventory']]
```

```python
groceries.loc[groceries['Subcategory']=='Snacks']
```

```python
groceries.sort_values('Subcategory',ascending=False)
```

```python
groceries.sort_values('Price_Dollars')
```

```python
groceries.sort_values(['Subcategory','Price_Dollars'])
```

```python
groceries[groceries['Low Inventory'] == 'Low Inventory'].sort_values('Price_Dollars')
```

```python
(groceries.groupby('Category')['Price_Dollars']
 .agg(['min','max','mean','nunique','count','sum'])
 .reset_index())
```

```python
groceries.groupby(['Category','Subcategory'])['Inventory'].agg(['min','max','mean','count']).reset_index()
```

```python
# INCORRECT
groceries.groupby("Category")[['Item','Price_Dollars']].max().reset_index()
```

```python
## CORRECT
(groceries[["Category",'Item','Price_Dollars']].sort_values("Price_Dollars",ascending=False)
.groupby('Category')
.head(1))
```

```python
happiness= pd.read_csv("./Data/happiness_survey_data.csv")
happiness.head()
```

```python
happiness[happiness.country_name == 'Mexico'].iloc[:,1:3].plot.line(x='year',y='happiness_score')
```

```python
(happiness[happiness.country_name.
 isin(['Canada','Mexico','United States'])].
loc[:,['country_name','year','happiness_score',]].pivot(index='year',columns='country_name',values='happiness_score').plot.line())
        
```
